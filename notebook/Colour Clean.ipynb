```python
import pandas as pd
import numpy as np
import colorsys as cs
```

```python
dfColor = pd.read_csv("listcolor.csv", index_col=None, header=None)
dfColor.head()
```

```python
dfColor[4] = dfColor[4].apply(lambda x: int(x, 16))
dfColor[2] = dfColor[2].apply(lambda x: int(x, 16))
dfColor[3] = dfColor[3].apply(lambda x: int(x, 16))
```

```python
for i, row in dfColor.iterrows():
    H, L, S = cs.rgb_to_hls(float(row[2])/255., float(row[3])/255., float(row[4])/255.)
    dfColor.loc[i,'Hue'],dfColor.loc[i,'Sat'],dfColor.loc[i,'Lum'] = int(H*360), int(S*100), int(L*100)
```

```python
dfColor.to_csv('listcolor.csv', index=False, header=False)
```
