```python
import pandas as pd
import numpy as np
```

```python
dfColor = pd.read_csv("../data/interim/Desc.csv")
dfColor.describe()
```

```python
dfColor.drop('index', inplace=True, axis=1)
```

```python
dfColor.head()
```

```python
dfColor1 = dfColor[:800]
dfColor2 = dfColor[700:1600]
dfColor3 = dfColor[1500:]
```

```python
dfColor1 = dfColor1.groupby('Desc2').mean().reset_index()
dfColor2 = dfColor2.groupby('Desc2').mean().reset_index()
dfColor3 = dfColor3.groupby('Desc2').mean().reset_index()
```

```python
dfColor12 = pd.concat([dfColor1, dfColor2])
dfColor23 = pd.concat([dfColor2, dfColor3])
```

```python
dfColorA = dfColor12.groupby('Desc2').mean().reset_index()
dfColorB = dfColor23.groupby('Desc2').mean().reset_index()
```

```python
dfColour = pd.concat([dfColorA, dfColorB])
```

```python
dfColour.drop_duplicates(inplace=True)
```

```python
dfColour.to_csv("Desc2.csv", index=None)
```
